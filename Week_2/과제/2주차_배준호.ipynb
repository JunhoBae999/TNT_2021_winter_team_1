```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pandas as pd
import numpy as np
import os

df = pd.read_csv("./mpg .csv")

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)


# df.replace([np.inf,-np.inf],np.nan).dropna(subset=["cylinders","displacement","horsepower","weight","acceleration","model_year"],how="all")
# df.where(pd.notnull(df), df.mean(), axis='columns')

df =df.fillna(df.mean())



# x = df[["cylinders","displacement","horsepower","weight","acceleration","model_year","origin","name"]]
x = df[["cylinders","displacement","horsepower","weight","acceleration","model_year"]]
y = df[["mpg"]]

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,test_size=0.2)


model = LinearRegression()
model.fit(x_train,y_train)

y_predict = model.predict(x_test)
print("r2 score : ", r2_score(y_test,y_predict))
print(model.score(x_train,y_train))

coefs = pd.DataFrame(zip(np.array(x_train.columns.values.tolist()),model.coef_[0]),columns = ['features', 'coefficients'])
coefs.reindex(coefs.coefficients.abs().sort_values(ascending=False).index)

"""
model_year, cylinders와는 강한 상관관계를 보이고 있고, horsepower, displacement,weight 와 약한 상관관계를 보이고 있다.
"""

coefs


```
